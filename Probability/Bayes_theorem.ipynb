## Task 1:- Find the box[20 points]

#### Suppose that we have three colored boxes r (red), b (blue), and g (green).
#### - Box r contains 3 apples, 4 oranges, and 3 limes, 
#### - Box b contains 1 apple, 1 orange, and 0 limes, and 
#### - Box g contains 3 apples, 3 oranges, and 4 limes. 
#### soln1: If a box is chosen at random with probabilities p(r) = 0.2, p(b) = 0.2, p(g) = 0.6, and a piece of fruit is removed from the box (with equal probability of selecting any of the items in the box), then what is the probability of selecting an apple? 
#### soln2: If we observe that the selected fruit is, in fact an orange, what is the probability that it came from the green box?

**Give your mathematical writing here**<br>
This problem can be solved using bayes theorem <br>
1) P(selecting an apple) = $P(apple|r) + P(apple|b) + P(apple|g) = (0.3\times0.2) + (0.5 \times0.2) +(0.3 \times 0.6) = 0.34$<br>
Therefore, P(selecting an apple) = 0.34<br><br>

2) P(Selected orange is from green box) = $P(green|orange)$ = $\frac{P(orange|green).P(green)}{P(orange)}$<br>
$P(orange) = P(orange|r)P(r) + P(orange|b)P(b) + P(orange|g) = (0.4\times 0.2) +(0.5\times 0.2) + (0.3\times 0.6) = 0.36$<br>
$P(green|orange) = \frac{0.3\times0.6}{0.36} = 0.5$<br>
    

```python
p_apple_n_r = 0.3
P_r = 0.2
p_apple_given_r = p_apple_n_r * P_r

p_apple_n_b = 0.5
p_b = 0.2
p_apple_given_b = p_apple_n_b *p_b

p_apple_n_g = 0.3
p_g = 0.6
p_apple_given_g = p_apple_n_g *p_g

P_apple = p_apple_given_r+p_apple_given_b+p_apple_given_g

soln1 = P_apple
print("P(selecting an apple)",soln1)

p_o_given_r = 0.4
p_r = 0.2
p_o_given_b = 0.5
p_b = 0.2
p_o_given_g = 0.3
p_g = 0.6
p_o = (p_o_given_r *p_r) + (p_o_given_b * p_b) + (p_o_given_g * p_g) 

p_g_given_o = (p_o_given_g*p_g)/(p_o)
soln2 = p_g_given_o
print("P(Selected orange is from green box",soln2)
```

## Task 2:- Let's get into magic[20 points]

#### Assume we have 3 cards mixed up in the hat which are identical in form but differs in the color, 
#### - the first card is colored green on both sides, 
#### - the second card is colored green on one side and blue on the other side 
#### - the third card is colored blue on both sides. 

#### Now one card is randomly selected, which has green color on one side. What is the probability that the other side is blue?

##### **Hint**: Monty Hall problem


Let C1 be the event that both sides of the card is coloured green<br>
C2 be the event that one side is green and other side blue<br>
C3 be the event that both sides of the card is coloured blue<br>
Let C be the event that the selected card has green on one side<br>
We need to find P(one side is green and other is blue)<br>
$P(C2|C) =\frac{P(C|C2)P(C2)}{(P(C|C1)P(C1))+(P(C|C2)P(C2))+P(C|C3)P(C3))}$ =$ \frac{\frac{1}{3}\frac{1}{2}}{\frac{1}{3} 1 + \frac{1}{3} \frac{1}{2} + \frac{1}{3} 0}$= $\frac{1}{3} = 0.333$<br>

```python
P_C1 = 1
P_C2 = 0.5
P_C3 = 0
P_C_given_C1 = 1/3
P_C_given_C2 = 1/3
P_C_given_C3 = 1/3
P_C2_given_C = (P_C_given_C2*P_C2)/((P_C_given_C1*P_C1)+(P_C_given_C2*P_C2)+(P_C_given_C3*P_C3))
soln3 = P_C2_given_C
print("P(Other side of selected card is blue is ",soln3)
```

## Task 3:- Gmail spam filter[20 points]

#### Gmail is used widely across the globe and the amount of Machine Learning incorporation is increasing periodically. One of the very useful incorporations is ham (non-spam)/spam classification. From past experience, we estimated that 50% of the emails in the inbox are spam. Gmail uses a Bayes classifier that filters out such spam before reaching your inbox. It has claimed that Gmail can detect 99% of spam emails, which accounts for 5% of false positive. If the upcoming mail is detected as spam, what is the probability that it is a ham?



**Answer**<br>
Let A be the event that an email is detected as spam<br>
Let B be the event that an email is spam<br>
Then B' is the event that an email is not spam<br>
From the question,<br>
$P(B) = P(B') = 0.5$<br>
$P(A|B) = 0.99$<br>
$P(A|B')= 0.05$<br>
We need to find $P(B'|A) = \frac{P(A|B')P(B')}{P(A|B)P(B)+P(A|B')P(B')}$
$P(B'|A)= \frac{0.05\times 0.5}{(0.99\times 0.5)(0.05\times 0.5)}= \frac{0.025}{0.52} = 0.481$

```python
P_b = 0.5
P_bc = 1-P_b
P_a_given_b = 0.99
P_a_given_bc = 0.05
P_bc_given_a = (P_a_given_bc*P_bc)/((P_a_given_b*P_b)+(P_a_given_bc*P_bc))
soln4 = P_bc_given_a
print("P(Upcoming email is ham",soln4)
```

## Task 4:- Gaussian Distribution[40 points]
#### Your final output should look like below 


#### Consider a Gaussian distribution $p_x(x)$ over x with mean μ = 6 and standard deviation $\sigma$ = 1. Plot this distribution.

```python
def gaussian(x, mu,sig):
    return (1./(np.sqrt(2.*np.pi)*sig)*np.exp(-np.power((x - mu)/sig, 2.)/2))

mu = 6
std =1

x = np.linspace(mu-6*std, mu+6*std)
px = gaussian(x,mu,sigma)
plt.plot(x,px,label='p(x)',color='g')
plt.legend()
plt.show()
```

#### Draw a sample of N = 50000 points from this distribution and plot a histogram of their values, which as expected agrees with the distribution $p_x(x)$. 

```python
s = np.random.normal(mu,sigma,50000)

#histogram
count, bins, ignored = plt.hist(s, 30, density=True,label='histogram',color='b')
plt.legend()

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='g',label='p(x)')
plt.legend()
plt.grid()
plt.show()
```

#### Now consider a non-linear change of variables from x to y given by  
#### $x=g(y)=ln(y)-ln(1-y)+5$  
#### The inverse of this function is given by  
#### $y=g^{-1}(x)=\frac{1}{1+exp(-x+5)}$  
#### which is a logistic sigmoid function. Plot this function along with the previous plot

```python
#sigmoid function
y = 1/(1+np.exp(-x+5))
plt.plot(x,y,label='y(x)',color='r')

s = np.random.normal(mu,sigma,50000)
#histogram
count, bins, ignored = plt.hist(s, 30, density=True,label='histogram',color='b')
plt.legend()

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='g',label='p(x)')
plt.legend()
plt.grid()
plt.show()
```

#### Plot the curve $p_x(g(y))$ that is a transformation of $p_x(x)$ as a function of x, $p_x(g(y))$ along with the previous plots, this should look exactly as the above one(reference provided above).

```python
#sigmoid function
y = 1/(1+np.exp(-x+5))
plt.plot(x,y,label='y(x)',color='r')

s = np.random.normal(mu,sigma,50000)

#histogram
count, bins, ignored = plt.hist(s, 30, density=True,label='histogram',color='b')
plt.legend()

#Fit line
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),linewidth=2, color='g',label='p(x)')
plt.legend()

#final plot

plt.plot(((np.log(y)- np.log(1-y)+5)*px),y,'--',color='c',label = 'p(g(y))')
plt.legend()
plt.grid()
plt.show()
```
